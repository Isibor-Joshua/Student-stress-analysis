# STUDENT STRESS FACTORS: A COMPREHENSIVE ANALYSIS

# Introduction

Stress is an emotional pressure resulting from the challenges, problems and difficulties of life and if not handled properly, may affect the person physically as well as mentally, ultimately deteriorating his performance in daily life activities. In this project, we are going to understand the underlying causes, their impact and correlations of stress factors on students.
The dataset used for this project is found in https://www.kaggle.com/datasets/rxnach/student-stress-factors-a-comprehensive-analysis.

### About the dataset

This dataset contains around 20 features that create the most impact on the Stress of a Student. The features are selected scientifically considering 5 major factors, they are Psychological, Physiological, Social, Environmental, and Academic Factors. Some of them are:

Psychological Factors: 'anxiety_level', 'self_esteem', 'mental_health_history', 'depression'.

Physiological Factors: 'headache', 'blood_pressure', 'sleep_quality', 'breathing_problem'.

Environmental Factors: 'noise_level', 'living_conditions', 'safety', 'basic_needs'.

Academic Factors: 'academic_performance', 'study_load', 'teacher_student_relationship', 'future_career_concerns'.

Social Factor: 'social_support', 'peer_pressure', 'extracurricular_activities', 'bullying'.

### Data Analysis

#### To get the number of students in the dataset

```python
number_of_students = len(df)
print("Number of students:", number_of_students)
```

#### To get the average anxiety level of the students

```python
Average_anxiety_level = int(round(df["anxiety_level"].mean()))

print("Average anxiety level:", Average_anxiety_level)

```

#### To know the number of students that have reported a history of mental health

```python
df["mental_health_history"].value_counts()
```

#### Graphical Representation

```python
mental_health_history = df["mental_health_history"]

mental_health_series = pd.Series(mental_health_history)
value_counts = mental_health_series.value_counts()

plt.figure(figsize=(4, 3))
value_counts.plot(kind='bar', color=['blue', 'green'])
plt.xlabel('Mental Health History')
plt.ylabel('Count')
plt.title('Mental Health History Report')
plt.xticks(rotation=0)
plt.show()
```

#### To know the number of students with  self-esteem below average

```python
self_esteem = (df.self_esteem < int(round(df.self_esteem.mean()))).sum() 
print("students with low self esteem:",self_esteem)
```

####  To get the percentage of those that have reported having depression

```python
depression_count = df['depression'].sum()
total_students = len(df.depression)
percentage_depression = (depression_count / total_students) 
print(f"percentage_depression: {percentage_depression:.2f}%")
```

####  To get the number of students with frequent headache

```python
headache = (df.headache > int(round(df.headache.mean())))).sum()
print("students with  frequent headache:",headache)
```

#### To get the average blood pressure among students

```python
Average_blood_pressure = int(round(df.blood_pressure.mean()))
print("Average blood presuure:",Average_blood_pressure)
```

####  To know the number of students with poor sleep

```python
sleep_quality = (df.sleep_quality < int(round(df.sleep_quality.mean()))).sum()
print("sleep quallity:",sleep_quality)
```

#### To get the sum of students with high noise level

```python
noise_level = (df.noise_level > int(round(df.noise_level.mean()))).sum()
print("high_noise_level:",noise_level)
```

#### Percentage of students with unsafe living condition

```python
living_condition_count=(df.living_conditions > int(round(df.noise_level.mean()))).sum()
total_students = len(df.living_conditions)
percentage_living_condition_count = (living_condition_count / total_students) * 100
print(f"percentage_living_condition_count: {percentage_living_condition_count:.2f}%")
```

#### Students with academic performance below average

```python
academic_performance = (df.academic_performance < int(round(df.academic_performance.mean()))).sum()
print("academic_performance_below_average:",academic_performance)
```

#### Average study load reported by students

```python
study_load = int(round(df.study_load.mean()))
print("Average_study_load:",study_load)
```

#### Students who have concern about their future careers

```python
concerns_about_career = df["future_career_concerns"]
students_with_concerns = sum(1 for concern in concerns_about_career if concern > 0)
print(f"Number of students with concerns about their future careers: {students_with_concerns}")
```

#### Students with strong support

```python
Strong_support = (df.social_support > int(round(df.social_support.mean()))).sum()
print("strong_support:",Strong_support)
```

#### Percentage of students who have experienced bullying

```python
bullying_count =(df.bullying > 0).sum()
total_students = len(df.bullying)
percentage_of_bullied_students = (bullying_count / total_students) * 100
print(f"percentage_of_bullied_students: {percentage_of_bullied_students:.2f}%")
```

#### Students who participate in extracurricular activities

```python
Students_with_extracurricular_activities = df["extracurricular_activities"]
Students_with_extracurricular_activities = sum(1 for concern in Students_with_extracurricular_activities if concern > 0)
print(f"Students with extracurricular activities: {Students_with_extracurricular_activities}")
```

### Correlation coefficients of various variables

#### Correlation between anxiety level and academic performance

```python
correlation_coefficient = df['anxiety_level'].corr(df['academic_performance'])
print("correlation coefficient:",correlation_coefficient)
```

#### Students with poor sleep quality also level of depression

```python
correlation_coefficient = df["sleep_quality"].corr(df["depression"])
print("correlation coefficient:", correlation_coefficient)
```

#### Are students who experience bullying more likely to have a history of mental health issues?

```python
correlation_coefficient = df["bullying"].corr(df["mental_health_history"])
print("correlation coefficient:", correlation_coefficient)
```

#### Heatmap

```python
df = pd.DataFrame(df, columns=['anxiety_level', 'self_esteem', 'mental_health_history', 'depression', 'headache','blood_pressure','sleep_quality','noise_level','living_conditions','safety','academic_performance','basic_needs','study_load','teacher_student_relationship','future_career_concerns','social_support','peer_pressure','extracurricular_activities','bullying','stress_level'])
correlation_matrix = df.corr()
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix, linewidths=1.5, linecolor='white', fmt='.1f',annot=True)
plt.title('Correlation Heatmap of Multiple Variables')
plt.show()
```

Summarily, a correlation coefficient is a number between -1 and 1 that tells you the strength and direction of a relationship between variables. In other words, it reflects how similar the measurements of two or more variables are across a dataset. The strength of relationship can be anywhere between −1 and +1. The stronger the correlation, the closer the correlation coefficient comes to ±1. If the coefficient is a positive number, the variables are directly related (i.e., as the value of one variable goes up, the value of the other also tends to do so) and if the  coefficient is a negative number, the variables are inversely related (i.e., as the value of one variable goes up, the value of the other tend to go down). This can be viewed on a heatmap.
